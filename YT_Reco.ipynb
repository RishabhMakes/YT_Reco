```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
print(type(soup))
```

```python
videos = soup.find_all(class_="ytd-rich-grid-video-renderer", attrs={"id": "video-title-link"})
# durations = soup("a")
print(len(videos))
for video in videos:
#     print(video)
    print(video.attrs['title'], video.attrs['href'])
#     print(video.child.attrs)
# print(durations[0].string)
```

```python
def parse_time(raw):
    pos=raw.index('ago')
#     print(pos, raw[pos])
    return(raw[(pos+1):-2])
    

for video in videos:
    
    raw = video.attrs['aria-label'].split()
    channel = video.parent.next_sibling.find(class_="style-scope ytd-channel-name complex-string").string
    
    print("Title: ", video.attrs['title'])
    print("Duration: ", parse_time(raw))
    print("Link: ", video.attrs['href'])
    print("Views: ",raw[-2])
    print("Channel: ",channel, "\n")
```
